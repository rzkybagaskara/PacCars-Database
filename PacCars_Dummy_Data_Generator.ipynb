### Importing Libraries

```python
import pandas as pd
import numpy as np
import random
```

### Global Faker Declaration

```python
from faker import Faker
from datetime import date
f = Faker('id_ID') #using Indonesian Locale
```

### Generate Primary Keys

```python
def generate_unique_ids(num_entries):
    return list(range(1, num_entries + 1)) #simpan ke dalam list

pengguna_ids = generate_unique_ids(40)
mobil_ids = generate_unique_ids(80)
iklan_ids = generate_unique_ids(100)
detail_iklan_ids = generate_unique_ids(100)
penawaran_ids = generate_unique_ids(50)
```

### Generate Data Pengguna

```python
def generate_data_pengguna(entries = 40):
  data = []
  for i in range(entries):
    random_date = f.date_between_dates(date_start=date(2020, 1, 1), date_end=date(2024, 6, 6))
    nama_depan = f.first_name()
    nama_belakang = f.last_name()
    nama = f"{nama_depan} {nama_belakang}" #menggunakan F-String untuk string interpolation (gabungan first dan last name)
    email = f"{nama.lower().replace(' ', '')}@{f.free_email_domain()}" #match email dengan nama dari pengguna
    fake_data_pengguna = {
        #"ID_Pengguna": np.random.randint(1, 100),
        "ID_Pengguna": pengguna_ids[i],
        "Nama": nama,
        "NoTelp": f.phone_number(),
        "Email": email,
        "Alamat": f.address(),
        "Domisili": f.city(),
        'Latitude': f.latitude(),
        'Longitude': f.longitude(),
        'Tanggal_Bergabung': random_date #set tahunnya antara 2020-2024
    }
    data.append(fake_data_pengguna)
  return pd.DataFrame(data)

generate_data_pengguna()
```

```python
generate_data_pengguna().to_excel('pengguna.xlsx', index=False)
```

```python
def generate_data_mobil(entries = 80):
  data = []
  jenis_lists = ['MPV', 'SUV', 'Sedan', 'Hatchback', 'Van', 'Station Wagon']
  years = [ 2005, 2010, 2011, 2020]
  transmisi_lists = ['manual', 'matic']

  for i in range(entries):
    merek = random.choice(list(merek_model_dict.keys())) #ubah jadi list kemudian ambil key dari dictionarnya (merek)
    model = random.choice(merek_model_dict[merek]) #ambil value berdasarkan key dict tersebut (merek)
    fake_data_mobil = {
        #"ID_Mobil": np.random.randint(1, 100),
        'ID_Mobil': mobil_ids[i],
        "Merek": merek,
        "Model": model,
        "Jenis": random.choice(jenis_lists),
        "Transmisi": random.choice(transmisi_lists),
        "Tahun": random.choice(years),
        "Odometer_KM": np.random.randint(8000, 200_000),
        "Warna": f.color_name(),
    }
    data.append(fake_data_mobil)
  return pd.DataFrame(data)

generate_data_mobil()
```

### Generate Data Penawaran

```python
def generate_data_penawaran(entries = 50):
  data = []
  status = ['Pending', 'Diterima', 'Ditolak']
  for i in range(entries):
    random_date = f.date_between_dates(date_start=date(2020, 1, 1), date_end=date(2024, 6, 6))
    fake_data_penawaran = {
        'ID_Penawaran': penawaran_ids[i],
        'ID_Pengguna': random.choice(pengguna_ids),
        "Harga_Penawaran": np.random.randint(70_000_000, 450_000_000),
        "Status_Penawaran": random.choice(status),
        'Tanggal_Penawaran': random_date
    }
    data.append(fake_data_penawaran)
  return pd.DataFrame(data)

generate_data_penawaran()
```
